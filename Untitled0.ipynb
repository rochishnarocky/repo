```python
#import necessary libaries
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/question_1.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')
```

```python
df['Total'] =df[['math score','reading score','writing score']].sum(axis=1)
```

```python
df.head()
```

```python
df['maximum'] =df[['math score','reading score','writing score']].max(axis=1)
```

```python
df.head()
```

```python
df['minimum'] =df[['math score','reading score','writing score']].min(axis=1)
```

```python
df.head()
```

```python
df['average'] =df[['math score','reading score','writing score']].mean(axis=1)
```

```python
import matplotlib.pyplot as plt


df['Total'].plot(kind='hist', bins=10, color='skyblue', edgecolour='black', title='histogram of total')

plt.xlabel
plt.ylabel('frequency')

plt.show()
```
